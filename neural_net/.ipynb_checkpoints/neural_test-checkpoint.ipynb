```python
class neural_network:
    def __init__(self,input_neurons,hidden_neurons,output_neurons,learning_rate):
        """
        ニューラルネットワークの初期化
        """
        self.inneurons=input_neurons #入力層のニューロン数
        self.hneurons=hidden_neurons #隠れ層のニューロン数
        self.oneurons=output_neurons #出力層のニューロン数
        self.lr=learning_rate #学習率
        self.weight_initializer()
    
    def weight_initialiser(self):
        self.w1=np.random.normal(
        0.0,
        pow(self.hneurons,-0.5),
        (self.oneurons,
        self.hneurons+1)
        )
        self.w2=np.random.normal(
                 0.0,
                 pow(self.hneurons,-0.5),
                 (self.oneurons,
                 self.hneurons+1)
                 )
        self.w_o=np.random.normal(
                0.0,
                pow(self.hneurons,-0.5),
                (self.oneurons,
                self.hneurons+1)
                )
        
    def sigmoid(self,x):#シグモイド
        return 1/(1+exp(-x))
    
    def softmax(self,x):
        c=np.max(x)
        exp_x=np.exp(x-c)
        sum_exp=np.sum(exp_x)
        y=exp_x/sum_exp_x
        return y
    
    def train(self,inputs_list,targets_list):
        """
        ニューラルネットワークの学習を行う
        ----
        inputs_list:訓練データの配列
        targets_list:正解ラベル
        """
        inputs=np.array(np.append(
                inputs_list,[1]),
                ndmin=2 #二次元化
                ).T
        ##隠れ層
        hidden_outputs=np.dot(
        self.w1, #隠れ層の重み
        inputs # 入力層の出力
        )
        
        hidden_outputs=self.sigmoid(hidden_inputs)
        hidden_outputs=np.append(
                        hidden_outputs
                        [[1]],
                        axis=0)
        #ここからP208[出力層]）
        
        
        
        
        
```

```python
input_neurons=10
hidden_neurons=10
output_neurons=10
learning_rate=0.1

n=neural_network(input_neurons,
                hidden_neurons,
                output_neurons,
                learning_rate)
```
