```python
import numpy as np
import matplotlib.pyplot as plt

def defferential(f,x):
    h=1e-4
    return (f(x+h)-f(x-h))/(2*h)


def function(x):
    return 0.01*x**2+0.1*x

def draw_line(f,x):
    dff=defferential(f,x)
    print(dff)
    y=f(x)-dff*x
    return lambda n: dff*n+y
    

x=np.arange(0.0,20.0,0.1)
y=function(x)

plt.xlabel("x")
plt.ylabel("f(x)")

tf=draw_line(function,6)
y2=tf(x)

plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```
