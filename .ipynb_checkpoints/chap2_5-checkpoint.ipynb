```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
#from mpl_toolkits.mplot3d import Axes3D

def func(x1,x2):
    return x1**2+x2**2


x1=np.arange(-3,3,0.25)
x2=np.arange(-3.3,0.25)
x1,x2=np.meshgrid(x1,x2)
y=func(x1,x2)

fig=plt.figure()
ax=Axes3D(fig)
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("f(x1,x2)")
ax.plot_wireframe(x1,x2,y)
plt.show()

```
